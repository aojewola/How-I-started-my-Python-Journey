```python
class Solution:
    def isHappy(n):
        if n==1:
            print(n)
        while (n!=1 and n!=4):
            c=[]
            m=[]
            d=[]
            d.extend(str(n))
            for l in d:
                c.append(int(l))
            for i in range(len(c)):
                kk=(c[i])*(c[i])
                m.append(kk)
            n=sum(m)
            if n==1:
                print(n)
            if n==4:
                print(n)
    isHappy(3)
```

```python
result = []
t = input("num: ")
for i in range(int(t)):
    N = input().split(" ")
    command = N[0]
    if command == 'insert':
       result.insert(int(N[1]), int(N[2]))
    if command == 'print':
        print(result)
    if command == 'remove':
       result.remove(int(N[1]))
    if command == 'append':
        result.append(int(N[1]))
    if command == 'reverse':
       result.reverse()
       print(result)
    if command == 'pop':
       result.pop()
    if command == 'sort':
       result.sort()
```
